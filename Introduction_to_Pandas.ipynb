```python
# Column from Series
seats = pd.Series([5, 5, 5, 5, 5, 5])

# New column called Seats
car_sales["Seats"] = seats
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
car_sales
```

```python
# Column from List
fuel_economy = [7.5, 9.2, 5.0, 9.6, 2.6]
car_sales["Fuel_per_100KM"] = fuel_economy
```

Series can have different length of values but the list should exactly match the number of values.

```python
fuel_economy = [7.5, 9.2, 5.0, 9.6, 2.6, 6.8, 4.2, 5.3, 5.8, 9.5]
car_sales["Fuel_per_100KM"] = fuel_economy
```

```python
car_sales
```

```python
car_sales["Total Fuel Used"] = ( car_sales["Odometer (KM)"] / 100 ) * car_sales["Fuel_per_100KM"]
```
